```python
import pandas as pd 

# 'skus' table
skus = pd.read_csv('./jddata/JD_sku_data.csv')
# 'users' table
users = pd.read_csv('./jddata/JD_user_data.csv')
# 'clicks' table
clicks = pd.read_csv('./jddata/JD_click_data.csv')
# 'orders' table
orders = pd.read_csv('./jddata/JD_order_data.csv')
# 'delivery' table
delivery = pd.read_csv('./jddata/JD_delivery_data.csv')
# 'inventory' table
inventory = pd.read_csv('./jddata/JD_inventory_data.csv')
# 'network' table
network = pd.read_csv('./jddata/JD_network_data.csv')
```

```python
network["dc_ori_region_id"] = network["region_ID"]
network["dc_ori"] = network["dc_ID"]
network["dc_des_region_id"] = network["region_ID"]
network["dc_des"] = network["dc_ID"] 

tmp = pd.merge(orders, users, on='user_ID', how='inner')[["order_ID", "quantity", "dc_ori", "dc_des", "user_ID", "education", "purchase_power", "city_level"]] 
tmp = pd.merge(tmp, network[["dc_ori_region_id", "dc_ori"]], on="dc_ori", how="inner") 
tmp = pd.merge(tmp, network[["dc_des_region_id", "dc_des"]], on="dc_des", how="inner")
tmp  
```

```python
# 9dc, 6dc, 8dc, 6dc, 9dc, 8dc, 5dc, 5dc
region_ids = [2, 3, 4, 5, 7, 9, 10, 24]
for region_id in region_ids:
    region = tmp[ (tmp["dc_des_region_id"] == region_id) & (tmp["city_level"] != -1) ]
    print(region["city_level"].mean()) 
```

```python
# region_id = 5
# 上海，昆山，南京，杭州，温州，义乌
dc_ids = [5, 13, 55, 54, 26, 65]
for dc_id in dc_ids:
    print("===============================")
    print(tmp[ (tmp["dc_des"] == dc_id) & (tmp["city_level"] != -1) ]["city_level"].mean())
```

```python
for dc_id in network["region_ID"].unique():
    print("===============================")
    print(tmp[ (tmp["dc_des"] == dc_id) & (tmp["city_level"] != -1) ]["city_level"].mean())
```

```python
# region_id = 5
# 上海，昆山，南京，杭州，温州，义乌
dc_ids = [5, 13, 55, 54, 26, 65]
for dc_ori in dc_ids:
    for dc_des in dc_ids:
        qua = tmp[ (tmp["dc_ori"] == dc_ori) & (tmp["dc_des"] == dc_des) ]["quantity"].sum()
        print(f"origin {dc_ori}, des {dc_des}: {qua}")
```
